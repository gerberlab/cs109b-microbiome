# Exploring the Vatanen dataset
In this notebook we load the relevant files from the Vatanen (2016) study. The study tracks the gut microbiomes of 117 children sampled over the first three years of life. 

```python
import pandas as pd
```

## Abundance data
The CSV file "abundance.csv" contains microbial abundance data obtained via 16s amplicon sequencing for each sample-ID.

```python
abundances = pd.read_csv("abundance.csv")
```

Let's see how the data is formatted

```python
abundances.head()
```

The sequence labels are given as columns, to make this easier to read, we do some reformatting. We set the sequences as rows and take the sampleIDs as column labels.

```python
abundances2 = abundances.transpose().rename(columns=abundances.transpose().iloc[0]).iloc[1:]
```

```python
abundances2.head()
```

This dataframe now has OTUs (represented by their sequences) as rows and samples as columns.

## Taxonomy
To see how the sequences map to taxonomy, we open the "dada2_placements.csv" file.

```python
dadaplace = pd.read_csv("dada2_placements.csv")
```

```python
dadaplace
```

This dataframe gives us the taxonomy for each sequence in the abundance dataframe.

## sample metadata
We next see how each sample maps to a subject and time point. The CSV file "sample_metadata.csv" specifies an associated subject ID and timepoint for each sample ID. 

```python
samplemeta = pd.read_csv("sample_metadata.csv", header=None)
```

```python
samplemeta.columns = ["sample_ID", "subjectID", "time"]
```

```python
samplemeta.head()
```

## subject metadata
The file "subject_data.csv" gives us a table showing how each host_subject_id maps to a label.

```python
subjdata = pd.read_csv("subject_data_augmented.csv")
```

```python
subjdata
```

The first column gives the host 'subjectID', same as that in the sample metadata file. The 'delivery' column gives the label for whether the delivery mode was vaginal or caesarean. The 'country' column gives the nationality (FIN=Finnish, RUS=Russian, EST=Estonian). Additional information is also given for infant gender, whether antibiotics were given in the first year (abx_first_year), and if they had various allergies. (See 'Vatanen_2016.pdf' in the references folder for more details).
